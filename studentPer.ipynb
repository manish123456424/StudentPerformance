```python
student['GAvg'] = (student['G1'] + student['G2'] + student['G3'])/ 3
```

Now lets create a grading based on its G Average:

Above 90% = Grade A

Between 70% & 90% = Grade B

Below 70% = Grade C

```python
def define_grade(df):
    grades = []
    
    for row in df['GAvg']:
        
        if row >= (0.9 * df['GAvg'].max()):
            grades.append('A')    
        elif row >= (0.7 * df['GAvg'].max()):
            grades.append('B')
        elif row < (0.7 * df['GAvg'].max()):
            grades.append('C')
    df['grades'] = grades
    return df
```

```python
student = define_grade(student)
student.head()
```

```python
mae=mean_absolute_error(ytest,y_pred)
mae
```

```python
#Random_Forest_Regression
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(xtrain,ytrain)
rfr_pred = rfr.predict(xtest)
print('MeanSq Error-:',mean_squared_error(ytest,rfr_pred))
print('R2 Square-:',r2_score(rfr_pred,ytest))
```

```python
#Decision_Tree_Regressor
from sklearn.tree import DecisionTreeRegressor
dsc = DecisionTreeRegressor(random_state=0)
dsc.fit(xtrain,ytrain)
dsc_pred = dsc.predict(xtest)
print('MeanSq Error-:',mean_squared_error(ytest,dsc_pred))
print('R2 Square-:',r2_score(dsc_pred,ytest))
```

```python
#KNN_Regressior
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(xtrain,ytrain)
knn_pred = knn.predict(xtest)
print('MeanSq Error-:',mean_squared_error(ytest,knn_pred))
print('R2 Square-:',r2_score(knn_pred,ytest))
```
